```python
import numpy as np
from tensorflow.keras.datasets import mnist
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
x_train=x_train.reshape(x_train.shape[0],784)/255.0
x_test=x_test.reshape(x_test.shape[0],784)/255.0

def one_hot(y):
  one_hoty=np.zeros((y.size,10))
  one_hoty[np.arange(y.size),y]=1
  return one_hoty

y_train_encod=one_hot(y_train)
y_test_encod=one_hot(y_test)
```

```python
def softmax(Z):
  expZ=np.exp(Z-np.max(Z, axis=1, keepdims=True))
  return expZ/expZ.sum(axis=1,keepdims=True)

def initialize_parameter():
  weights=np.random.randn(784,10)*.01
  biases=np.zeros((1,10))
  return weights,biases

def forward(X,weight,biases):
  Z=np.dot(X,weight)+biases
  return softmax(Z)

def update_parameters(X,A,Y,weights,biases,learn_rate):
  m=X.shape[0]
  error=A-Y

  dw=(1/m)*np.dot(X.T,error)
  db=(1/m)*np.sum(error,axis=0,keepdims=True)

  weights-=learn_rate*dw
  biases-=learn_rate*db
  return weights,biases
```

```python
weights,biases=initialize_parameter()
iterate=100
learn_rate=1.0
for epo in range(iterate):
  prediction=forward(x_train,weights,biases)
  weights,biases=update_parameters(x_train,prediction,y_train_encod,weights,biases,learn_rate)

  predic_class=np.argmax(prediction,axis=1)
  accuracy=np.mean(predic_class==y_train)*100
  print(f"Epoch: {epo+1}/{iterate}-accuracy:{accuracy:.2f}% ")
```

```python
test_predictions = forward(x_test, weights, biases)
test_classes = np.argmax(test_predictions, axis=1)
test_accuracy = np.mean(test_classes == y_test) * 100
print(f"Final Accuracy on Hidden Test Images: {test_accuracy:.2f}%")
```
