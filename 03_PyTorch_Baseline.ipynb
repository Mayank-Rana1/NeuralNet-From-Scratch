```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader

# 1. SETUP THE CONVEYOR BELT (Data)
transform = transforms.ToTensor()
train_data = datasets.MNIST(root='./data', train=True, download=True, transform=transform)
test_data = datasets.MNIST(root='./data', train=False, download=True, transform=transform)

train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
test_loader = DataLoader(test_data, batch_size=64, shuffle=False)

# 2. BUILD THE ROBOT (Model)
class DeepBrain(nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(784, 128)
        self.layer2 = nn.Linear(128, 10)

    def forward(self, x):
        x = x.view(-1, 28 * 28)           # Flatten
        x = torch.relu(self.layer1(x))    # Math + Bend
        x = self.layer2(x)                # Final Guesses
        return x

model = DeepBrain()

# 3. THE GRADER AND THE MECHANIC
criterion = nn.CrossEntropyLoss()
optimizer = optim.SGD(model.parameters(), lr=0.1)

# 4. THE TRAINING LOOP
epochs = 10
for epoch in range(epochs):
    model.train()
    correct = 0
    total = 0

    for images, labels in train_loader:
        optimizer.zero_grad()             # Sweep the floor
        outputs = model(images)           # Guess
        loss = criterion(outputs, labels) # Check error
        loss.backward()                   # Calculate blame
        optimizer.step()                  # Tweak weights

        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    print(f"Epoch {epoch+1} - Accuracy: {100 * correct / total:.2f}%")

# 5. THE TEST LOOP
model.eval()
correct = 0
total = 0
with torch.no_grad(): # Turn off calculus to save memory
    for images, labels in test_loader:
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Final Test Accuracy: {100 * correct / total:.2f}%")
```
